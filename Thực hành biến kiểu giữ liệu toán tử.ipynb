```python
# vẽ mặt cười 

import turtle 
pen = turtle.Turtle()
pen.fillcolor("blue")
pen.pensize(5)
pen.pencolor("red") #doi mau net ve 
pen.shape("circle")

pen.penup()

pen.goto(0,-200)

facesize = 200 

pen.pendown()
pen.begin_fill()
pen.circle(facesize)
pen


#ve mat 1 
pen.penup()
pen.goto(-50,100)
yey1 = 15

pen.pendown()
pen.begin_fill()
pen.circle(yey1)
pen.end_fill()


#ve mat 1 
pen.penup()
pen.goto(50,100)
yey2= 15

pen.pendown()
pen.begin_fill()
pen.circle(yey2)
pen.end_fill()

#ve mui
pen.penup()
pen.goto(0,50)
yey2= 15
nose = 50
pen.pendown()
pen.circle(-nose,steps=3)

#ve mieng cuoi

pen.penup()
pen.goto(-100, -70)

pen.pendown()
pen.right(90)
pen.circle(100,180)






turtle.done()



```
